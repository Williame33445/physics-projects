# Density functional theory for a Cu crystal

This notebook 

considers some simple examples with the ASE library and then uses density functional theory on various systems.

Work is inspired from: https://kitchingroup.cheme.cmu.edu/dft-book/dft.html#orgheadline11

```python
from ase import Atoms, Atom
from ase.io import write
from ase.visualize import view
from ase.build import molecule
from ase.calculators.emt import EMT
import numpy as np
import matplotlib.pyplot as plt


dir = "C:\workspace\git-repos\physics-projects\electronic_structure\density-functional-theory\images"
```

Explain what EMT is

```python

aS = np.arange(3.1,4.5,0.01)
Es = np.empty(len(aS))

for i,a in enumerate(aS):
    atoms = Atoms([Atom("Cu", [0, 0, 0]), Atom("Cu", [a/2, 0, a/2]),
                  Atom("Cu", [a/2, a/2, 0]),Atom("Cu", [0, a/2, a/2])],
                  cell=(a,a,a),pbc=True,calculator=EMT())
    Es[i] = atoms.get_total_energy()

plt.plot(aS,Es)
plt.title("Energy vs lattice parameter for Cu")
plt.ylabel("Energy (eV)")
plt.xlabel("lattice parameter (Ang)")
plt.show()
```
